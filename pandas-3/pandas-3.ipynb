### 4.What is the maximum and minimum Number of contracts traded during the 2009

```python
year = data.loc[data["year"] == 2009,["volume","year"]]
max_ = year["volume"].max()
min_ = year["volume"].min()
print("max number of contracts traded during the 2009: ",max_)
print("\nmin number of contracts traded during the 2009: ",min_)
```

### 5. Which year have  maximum and minimum Number of contracts traded

```python
print('Minimum',data.loc[data['volume']==data['volume'].min(),['date']])
```

### 8.Which year corn have the highest and lowest Number of contracts traded


```python
c=data.loc[data['commodity']=='Corn']
print('corn having lowest Number of contracts traded :',c.loc[c['volume'].idxmin(),'year'])
print('corn having highest Number of contracts traded :',c.loc[c['volume'].idxmax(),'year'])
```
