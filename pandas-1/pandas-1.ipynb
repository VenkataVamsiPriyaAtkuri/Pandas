### 2.What is the average ram size of all the collected data

```python
avg=data['ram(GB)'].mean()
print('The average ram size of all the collected data : ',avg)
```

### 3. Which phone is having the maximum battery capacity

```python
data.loc[data['battery']==data['battery'].max(),['phone_name','brand','battery']]
```

### 4.What is the maximum battery capacity provided by Huawei Company

```python
h=data.loc[data['brand']=='Huawei']
c=h['battery'].max()
print('Max battery capacity provided by Huawei company is :',c)
```

### 5. Which Phone brand is providing the highest and lowest ram

```python
max_=data.loc[data['ram(GB)']==data['ram(GB)'].max(),['phone_name','brand','ram(GB)']]
min_=data.loc[data['ram(GB)']==data['ram(GB)'].min(),['phone_name','brand','ram(GB)']]
```
